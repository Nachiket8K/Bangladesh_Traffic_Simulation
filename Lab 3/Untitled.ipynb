```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

if __name__ == '__main__':
    G = nx.DiGraph()
    G.add_node(1)
    G.add_nodes_from([2, 3, 4, 5])
    G.add_weighted_edges_from([(1, 2, 2.5),(2, 3, 1.2),(2, 3, 3),(3, 4, 4),(2, 5, 1),(4, 5, 2.3),(1, 4, 4.6)])



    nx.draw(G,with_labels=True,node_color='orange')
    plt.show
```

```python

if __name__ == '__main__':
    G = nx.DiGraph()

    G.add_nodes_from([2, 3, 4, 5])
    G.add_weighted_edges_from([(3, 2, 2.5),(2, 3, 1.2),(2, 5, 3),(4, 5, 4)])

    pos= nx.spring_layout(G)
    nx.draw(G, pos , with_labels=True, node_color='orange')
    labels=nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
    nx.draw(G,with_labels=True,node_color='orange')
    plt.show
```
